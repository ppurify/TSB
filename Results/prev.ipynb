```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
```

```python
def format_coord(_data_df):
    for p in range(len(_data_df)):
    # for points in prev_Truck_10_shortest_data_df['Pick_station']:

        pick_str = _data_df['Pick_station'][p].strip("()")  # Remove parentheses
        pick_str = pick_str.split(", ")  # Split by comma and space

        pick_x = int(pick_str[0])
        pick_z = int(pick_str[2])
        
        drop_str = _data_df['Drop_station'][p].strip("()")  # Remove parentheses
        drop_str = drop_str.split(", ")  # Split by comma and space
        
        drop_x = int(drop_str[0])
        drop_z = int(drop_str[2])
        
        # formatted_coordinates_pick = f"({pick_x}, {pick_z})"
        # formatted_coordinates_drop = f"({drop_x}, {drop_z})"
        
        formatted_coordinates_pick = (pick_x, pick_z)
        formatted_coordinates_drop = (drop_x, drop_z)
        
        
        _data_df['Pick_station'][p] = formatted_coordinates_pick
        _data_df['Drop_station'][p] = formatted_coordinates_drop
        
    return _data_df
```

```python
def draw_grid(_grid_length, _grid_height, _square_size, _coordinates_list_1, _coordinates_list_2):
    # Create figure and axis
    fig, ax = plt.subplots(figsize=(_grid_length + _square_size, _grid_height + _square_size))
    # Set aspect ratio to equal, so squares appear as squares
    ax.set_aspect('equal')

    # Loop through rows and columns to draw squares
    for row in range(_grid_height):
        for col in range(_grid_length):
            x = col * _square_size - _square_size / 2
            y = row * _square_size - _square_size / 2

            # Draw the grid square
            ax.add_patch(patches.Rectangle((x, y), _square_size, _square_size, linewidth=1, edgecolor='black', facecolor='none'))

    # Set axis limits
    ax.set_xlim(-_square_size / 2, _grid_length * _square_size - _square_size / 2)
    ax.set_ylim(-_square_size / 2, _grid_height * _square_size - _square_size / 2)

    # Add numbers to x-axis and y-axis
    plt.xticks(range(0, _grid_length * _square_size, _square_size))
    plt.yticks(range(0, _grid_height * _square_size, _square_size))

    # Add labels to x-axis and y-axis
    plt.xlabel("X")
    plt.ylabel("Y")

    for coord_x, coord_y in _coordinates_list_1:
        plt.scatter(coord_x, coord_y, color='red', marker='o', s=300, zorder=2, alpha= 0.3)  # zorder to plot on top of grid
    
    for coord_x, coord_y in _coordinates_list_2:
        plt.scatter(coord_x, coord_y, color='blue', marker='*', s=200, zorder=2, alpha=0.3)

    plt.show()
```

```python
def calculate_inbound_outbound_count(_grid_height, _square_size, _pick_list):
    grid_height_y_max = (_grid_height-1) * _square_size

    outbound_count = 0
    inbound_count = 0

    for i in _pick_list:
        if i[1] < grid_height_y_max:
            outbound_count += 1
        
        else:
            inbound_count += 1
            
    print("outbound_count =", outbound_count)
    print("inbound_count =", inbound_count)
```

```python
# Parameters
grid_length = 31
grid_height = 9

block_length = 9
block_height = 1

block_num_in_row = 3
block_num_in_col = 4

square_size = 25
```

### prev_Truck_10_shortest

```python
data_file = 'prev/shortest/prev_Truck_10_shortest.csv'

prev_Truck_10_shortest_data = pd.read_csv(data_file, header=None)

data = prev_Truck_10_shortest_data.iloc[1:].values

column_names = prev_Truck_10_shortest_data.iloc[0].tolist()

prev_Truck_10_shortest_data_df = pd.DataFrame(data, columns=column_names)

prev_Truck_10_shortest_data_df = format_coord(prev_Truck_10_shortest_data_df)

prev_Truck_10_shortest_pick_list = prev_Truck_10_shortest_data_df['Pick_station'].tolist()
prev_Truck_10_shortest_drop_list = prev_Truck_10_shortest_data_df['Drop_station'].tolist()

draw_grid(grid_length, grid_height, square_size, prev_Truck_10_shortest_pick_list, prev_Truck_10_shortest_drop_list)
calculate_inbound_outbound_count(grid_height, square_size, prev_Truck_10_shortest_pick_list)
```
